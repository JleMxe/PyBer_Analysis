### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])
# Display the data table for preview
pyber_data_df.head(10)
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_rides_by_type = pyber_data_df.groupby(["type"])["ride_id"].count()
total_rides_by_type
```

```python
# 2. Get the total drivers for each city type
#total_drivers_by_type = pyber_data_df.groupby(["type"])["driver_count"].sum()
total_drivers_by_type = city_data_df.groupby(["type"])["driver_count"].sum()
total_drivers_by_type
```

```python
#  3. Get the total amount of fares for each city type
total_fares_by_type = pyber_data_df.groupby(["type"])["fare"].sum()
total_fares_by_type
```

```python
#  4. Get the average fare per ride for each city type. 
average_fare_per_ride_by_type = total_fares_by_type / total_rides_by_type
average_fare_per_ride_by_type
```

```python
# 5. Get the average fare per driver for each city type. 
average_fare_per_driver_by_type = total_fares_by_type / total_drivers_by_type
average_fare_per_driver_by_type
```

```python
#  6. Create a PyBer summary DataFrame.
# Sytax for concat found at : https://www.geeksforgeeks.org/combine-two-pandas-series-into-a-dataframe/
pyber_summary_df = {"Total Rides":total_rides_by_type, 
                    "Total Drivers":total_drivers_by_type, 
                    "Total Fares":total_fares_by_type,
                    "Average Fare per Ride":average_fare_per_ride_by_type, 
                    "Average Fare per Driver":average_fare_per_driver_by_type}
pyber_summary_df = pd.concat(pyber_summary_df,axis =1)
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
#  8. Format the columns.
# Sytax for map found at: https://stackoverflow.com/questions/36604384/output-of-column-in-pandas-dataframe-from-float-to-currency-negative-values
pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("${:,.2f}".format)
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
# 1. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
total_fares_by_type_and_date_df = pyber_data_df.groupby(["type","date"]).sum()[["fare"]]
total_fares_by_type_and_date_df
```

```python
# 2. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
total_fares_by_type_and_date_df = total_fares_by_type_and_date_df.reset_index()
```

```python
# 3. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date.

# pivot syntax found at: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.pivot.html
total_fares_by_type_and_date_pivot = total_fares_by_type_and_date_df.pivot(index="date", columns="type", values="fare")
total_fares_by_type_and_date_pivot
```

```python
# 4. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
#Syntax for loc found at: https://trenton3983.github.io/files/projects/2019-02-04_manipulating_dataframes_with_pandas/2019-02-04_manipulating_dataframes_with_pandas.html
total_fares_by_type_and_date_pivot_filtered_df = total_fares_by_type_and_date_pivot.loc["2019-01-01":"2019-04-28"]
total_fares_by_type_and_date_pivot_filtered_df.head()
```

```python
# 5. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
total_fares_by_type_and_date_pivot_filtered_df.index = pd.to_datetime(total_fares_by_type_and_date_pivot_filtered_df.index)
```

```python
# 6. Check that the datatype for the index is datetime using df.info()
total_fares_by_type_and_date_pivot_filtered_df.info()
```

```python
# 7. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
total_fares_by_type_and_week_filtered_df = total_fares_by_type_and_date_pivot_filtered_df.resample("W").sum()
total_fares_by_type_and_week_filtered_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
# changing the rc parameters and plotting a line plot
#Syntax to adjust chart size: https://www.geeksforgeeks.org/change-plot-size-in-matplotlib-python/
plt.rcParams['figure.figsize'] = [9, 3]
total_fares_by_type_and_week_filtered_df.plot()
# Legend location syntax found at: https://matplotlib.org/3.5.0/tutorials/intermediate/legend_guide.html
plt.legend(loc="center")
# Title and axis label syntax: https://python-graph-gallery.com/4-add-title-and-axis-label
plt.title("Total Fare by City Type")
plt.ylabel("Fare ($USD)")
plt.xlabel("")
#Save chart
# Resolution fo rcutting off the bottom: https://www.youtube.com/watch?v=C8MT-A7Mvk4
plt.savefig("analysis/PyBer_fare_summary.png",bbox_inches='tight')
```
