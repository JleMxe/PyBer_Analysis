```python
# Import Dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot # This is the MatLab method.
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() # This is th eobject oriented method (preferred)
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() # Another way of writing it (still object oriented)
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot. # Showing the use of plt.show() Use this ONLY ONCE per session. If it messes up, restart the kernel
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Set the x limit between Jan and June
plt.xlim("Jan", "June")
# plt.xlim(0, 5) # This format also works
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
# Change the figure to 20 pixels wide by 10 pixels high
plt.figure(figsize=(20,10))
#plt.subplots(figsize=(20, 10)) This format works too
```

## Scatter PLot MATLAB Method

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
## The 'o' tells it to be a scatter plot instead of a line chart
plt.plot(x_axis, y_axis, 'o')
```

```python
## Same result just different method
plt.scatter(x_axis, y_axis)
```

```python
## Same result just different method
plt.scatter(x_axis, y_axis, color = 'red', label = "Chicago")
```
